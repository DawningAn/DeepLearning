```python
#根据正则化，实现岭回归模型
import numpy as np
import matplotlib.pyplot as plt
```

```python
#读入数据
data = np.genfromtxt("longley.csv", delimiter=',')
#切分数据
x_data = data[1:, 2:]
#这里要y_data增加维度
y_data = data[1:, 1, np.newaxis]
print(x_data)
print(y_data)
```

```python
print(np.mat(x_data).shape)
print(np.mat(y_data).shape)
#加入偏置，将b吸入向量形式
print(np.ones((16, 1)))  # 16行1列的全1.矩阵
# np.concatenate((np.ones((16,1)), np.zeros((16,2))),axis=1)    # 演示np.concatenate的用法
X_data = np.concatenate((np.ones((16, 1)), x_data), axis=1)
X_data
```

```python
# 岭回归标准方程法求参数
def weights(xArr, yArr, lam):
    xMat = np.mat(xArr)
    yMat = np.mat(yArr)
    xTx = xMat.T * xMat  # 矩阵乘法
    rxTx = xTx + lam*np.eye(xMat.shape[1])  # 生成7行7列的单位矩阵
    if np.linalg.det(rxTx) == 0.0:  # 如果矩阵值为零，就结束
        print("This matrix cannot do inverse")
        return
    # 求逆
    w = rxTx.I * xMat.T * yMat
    return w
```

```python
w = weights(X_data,y_data,0.2)
print(w)  # 得到7个系数
```

```python
# 预测结果
np.mat(X_data) * np.mat(w)
```
