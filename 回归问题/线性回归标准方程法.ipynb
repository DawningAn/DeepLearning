```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#还是以data这个二维数据集示例
data = np.genfromtxt("data.csv", delimiter=',')
x_data = data[:, 0, np.newaxis]
y_data = data[:, 1, np.newaxis]
plt.scatter(x_data, y_data)
print(x_data)
print(y_data)
```

```python
print(np.mat(x_data).shape)  #将向量格式转换成矩阵
print(np.mat(y_data).shape)  #100行1列
# 进行数组合并（相当于将偏置b合并到W矩阵里）
X_data = np.concatenate((np.ones((100, 1)), x_data), axis=1)  #生成100行1列的全1矩阵
print(X_data.shape)
print(X_data)  #就是把1和x_data进行矩阵合并
```

```python
#标准方程求参数
def weights(xArr, yArr):
    #将传进来的xdata，ydata转换为矩阵
    xMat = np.mat(xArr)
    yMat = np.mat(yArr)
    xTx = xMat.T * xMat  #矩阵乘法
    #计算矩阵的值，如果为零，则没有逆矩阵
    if np.linalg.det(xTx) == 0.0:
        print("This matix cannot do inverse")
        return
    #xTx.I为xTx的逆矩阵
    w = xTx.I * xMat.T * yMat
    return w
```

```python
w = weights(X_data, y_data)
print(w)
```

```python
x_test = np.array([[30],[71]])
y_test = w[0] + x_test * w[1]
plt.plot(x_data,y_data,'g.')
plt.plot(x_test,y_test,'r')
plt.show()
```
