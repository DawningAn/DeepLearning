```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

在一般情况下，我们更多的取使用已有的模型库来训练，辅助我们得到结果。

```python
# 载入数据
data = np.genfromtxt("data.csv", delimiter=",")
x_data = data[:,0]  # 得到数据集的x
y_data = data[:,1]  # 得到数据集的y
#设置图形大小
plt.figure(figsize=(20,10),dpi=80)
plt.scatter(x_data,y_data)
plt.show()
print(x_data.shape)  # (100,) 即该数据集样本点有100个
```

```python
x_data = data[:,0,np.newaxis]  #扩充维数（这是因为sklearn里的线性回归模型要求数据必须是不低于二维的）
y_data = data[:,1,np.newaxis]
# 创建并拟合模型
model = LinearRegression()
model.fit(x_data, y_data)
```

```python
# 画图
plt.plot(x_data, y_data,'b.')  #第三个参数为'格式控制字符串','b.'表示蓝色点
plt.plot(x_data, model.predict(x_data), 'r')
plt.show()
```
